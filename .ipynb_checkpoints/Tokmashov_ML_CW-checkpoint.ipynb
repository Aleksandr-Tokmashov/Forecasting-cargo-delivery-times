## Подготовка данных

### Удаление константных столбцов

```python
# Удаление константных столбцов
constant_cols = [col for col in df.columns if df[col].nunique() == 1]
df = df.drop(columns=constant_cols)
```

### Обработка пропусков

```python
# Обработка пропусков
missing_percent = df.isnull().mean().sort_values(ascending=False) * 100
cols_to_drop = missing_percent[missing_percent > 80].index
df = df.drop(columns=cols_to_drop)

# Заполнение пропусков
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

cat_cols = df.select_dtypes(include=['object', 'category']).columns
df[cat_cols] = df[cat_cols].fillna('Unknown')
```

### Обработка дат

```python
# Временные признаки
df['Дата и время начала рейса'] = pd.to_datetime(df['Дата и время начала рейса'])
df['Дата и время окончания рейса'] = pd.to_datetime(df['Дата и время окончания рейса'])
```

### Создание новых признаков

```python
# Кодируем категориальные признаки
from sklearn.preprocessing import LabelEncoder
cat_cols = []
for col in ['Государство отправления', 'Государство назначения', 'Наименование груза',
            'Операция', 'Состояние вагона',
            ' Вид планового ремонта', ' Вид ремонта']:
    if col in df.columns:
        df[col] = LabelEncoder().fit_transform(df[col].astype(str))
        cat_cols.append(col)
```

```python
# Базовые временные признаки
df['День недели'] = df['Дата и время начала рейса'].dt.day_of_week
df['Месяц'] = df['Дата и время начала рейса'].dt.month
df['Длительность рейса (часы)'] = (df['Дата и время окончания рейса'] - df['Дата и время начала рейса']).dt.total_seconds() / 3600
```

```python
# Основные числовые признаки
numeric_features = [
    'Пробег общий (км)',
    'Вес груза (кг)',
    'Расстояние общее (км)',
    'День недели',
    'Месяц',
    'Государство отправления',
    'Наименование груза',
    'Операция'
]
```

```python
# Расчетные показатели
df['Грузо-пробег'] = df['Вес груза (кг)'] * df['Пробег общий (км)']
df['Грузоплотность'] = df['Вес груза (кг)'] / (df['Пробег общий (км)'] + 1)
df['Скорость доставки'] = df['Расстояние общее (км)'] / df['Срок доставки факт']

numeric_features += [
    'Грузо-пробег',
    'Грузоплотность',
    'Скорость доставки'
]
```

```python
# Целевая переменная
target = 'Срок доставки факт'
```

```python
# Удаление строк с пропусками в ключевых столбцах
initial_count = len(df)
df = df.dropna(subset=numeric_features + [target])
cleaned_count = len(df)
print(f"Удалено записей: {initial_count - cleaned_count}")
```

```python
# Фильтрация аномальных значений
df = df[(df['Срок доставки факт'] > 0) & 
        (df['Срок доставки факт'] <= 50) &  
        (df['Расстояние общее (км)'] > 0)]

df = df.drop(columns=['Дата и время окончания рейса'])
```

```python
# Дополнительные временные признаки
df['Год'] = df['Дата и время начала рейса'].dt.year
df['Неделя года'] = df['Дата и время начала рейса'].dt.isocalendar().week
df['День месяца'] = df['Дата и время начала рейса'].dt.day
df['Часть месяца'] = pd.cut(df['День месяца'], 
                           bins=[1, 10, 20, 31],
                           labels=['Начало', 'Середина', 'Конец'])
```

```python
# Статистические признаки
country_weight_mean = df.groupby('Государство отправления')['Вес груза (кг)'].transform('mean')
df['Отклонение веса от среднего по стране'] = df['Вес груза (кг)'] - country_weight_mean
```

```python
# Преобразования признаков
df['log_пробег'] = np.log1p(df['Пробег общий (км)'])
df['sqrt_расстояние'] = np.sqrt(df['Расстояние общее (км)'])
```

```python
# Дополнительные расчетные признаки
df['Время_на_единицу_веса'] = df['Срок доставки факт'] / (df['Вес груза (кг)'] + 1e-6)
```

```python
# Категоризация расстояний
df['Категория_расстояния'] = pd.qcut(
    df['Расстояние общее (км)'],
    q=5,
    labels=['оч.короткое', 'короткое', 'среднее', 'длинное', 'оч.длинное']
)
```

```python
# Сезонность
seasons = {
    1: 'Зима', 2: 'Зима', 3: 'Весна',
    4: 'Весна', 5: 'Весна', 6: 'Лето',
    7: 'Лето', 8: 'Лето', 9: 'Осень',
    10: 'Осень', 11: 'Осень', 12: 'Зима'
}
df['Сезон'] = df['Месяц'].map(seasons)
```

```python
# Праздничные дни
holidays = ['2024-01-01', '2024-01-07', '2024-02-23']
df['Праздничный день'] = df['Дата и время начала рейса'].dt.date.isin(pd.to_datetime(holidays).date)
```

```python
# Временные лаги
df = df.sort_values('Дата и время начала рейса')
df['Средний срок за последние 7 дней'] = (
    df
    .set_index('Дата и время начала рейса')
    ['Срок доставки факт']
    .rolling('7D')
    .mean()
    .values
)
```

### Подготовка финального набора признаков

```python
final_features = [col for col in numeric_features if col not in categorical_features]

# Разделение данных
X = df[final_features]
y = df[target]

# Создание пайплайна предобработки
preprocessor = Pipeline(steps=[
    ('scaler', StandardScaler())
])

# Разделение на train/test
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=42,
)

# Применение преобразований
X_train_preprocessed = preprocessor.fit_transform(X_train)
X_test_preprocessed = preprocessor.transform(X_test)
```

### Обучение AutoML

```python
# Инициализация H2O
h2o.init(max_mem_size='4G')

# Подготовка данных для H2O
def prepare_h2o_data(X, y):
    df = pd.concat([X, y], axis=1)
    df = df.dropna(subset=[y.name])
    numeric_cols = df.select_dtypes(include=np.number).columns
    df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())
    return h2o.H2OFrame(df)

try:
    h2o_train = prepare_h2o_data(X_train, y_train)
    h2o_test = prepare_h2o_data(X_test, y_test)
    
    h2o_train['Срок доставки факт'] = h2o_train['Срок доставки факт'].asnumeric()
    h2o_test['Срок доставки факт'] = h2o_test['Срок доставки факт'].asnumeric()
    
    # Запуск AutoML
    aml = H2OAutoML(
        max_models=5,
        max_runtime_secs=600,
        include_algos=["GLM", "DRF"],
        seed=42,
    )
    
    aml.train(
        x=list(X_train.columns),
        y='Срок доставки факт',
        training_frame=h2o_train,
        validation_frame=h2o_test
    )
    
    # Прогнозирование и оценка
    if aml.leaderboard.nrows > 0:
        best_model = aml.leader
        preds = best_model.predict(h2o_test)
        
        y_pred = preds['predict'].as_data_frame().values.flatten()
        y_true = h2o_test['Срок доставки факт'].as_data_frame().values.flatten()
        
        mask = ~np.isnan(y_pred) & ~np.isnan(y_true)
        y_pred_clean = y_pred[mask]
        y_true_clean = y_true[mask]
        
        print("\nОсновные метрики регрессии:")
        print(f"MSE: {mean_squared_error(y_true_clean, y_pred_clean):.4f}")
        print(f"MAE: {mean_absolute_error(y_true_clean, y_pred_clean):.4f}")
        print(f"R2 Score: {r2_score(y_true_clean, y_pred_clean):.4f}")

except Exception as e:
    print(f"Ошибка: {e}")
finally:
    h2o.shutdown()
```

### Обучение собственной модели

```python
# Получение названий числовых и категориальных признаков
num_cols = X.select_dtypes(include=['int64', 'float64']).columns.tolist()
cat_cols = X.select_dtypes(include=['object', 'category']).columns.tolist()

# Удаление лишних столбцов
columns_to_drop = ['Номер вагона', 'Номер накладной']
num_cols = [col for col in num_cols if col not in columns_to_drop]
cat_cols = [col for col in cat_cols if col not in columns_to_drop]

# Создание преобразователя
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), num_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore'), cat_cols)
    ])

X_processed = preprocessor.fit_transform(X)

# Разделение данных
X_train, X_test, y_train, y_test = train_test_split(
    X_processed, y, test_size=0.3, random_state=42
)
```

```python
xgb_model = XGBRegressor(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=6,
    subsample=0.9,
    colsample_bytree=0.8,
    random_state=42
)
```

```python
cat_model = CatBoostRegressor(
    iterations=300,
    learning_rate=0.05,
    depth=6,
    l2_leaf_reg=3,
    random_seed=42,
    silent=True
)
```

```python
lgbm_model = lgb.LGBMRegressor(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=6,
    subsample=0.9,
    colsample_bytree=0.8,
    random_state=42
)
```

```python
# Список моделей для стекинга
estimators = [
    ('lgbm', lgbm_model),
    ('xgb', xgb_model),
    ('cat', cat_model)
]
```

```python
# Используем RidgeCV с фиксированными альфами
meta_model = RidgeCV(alphas=[0.1, 1.0, 10.0], cv=5)

stacking_model = StackingRegressor(
    estimators=estimators,
    final_estimator=meta_model,
    cv=5,
    n_jobs=1
)

stacking_model.fit(X_train, y_train)

# 4. Предсказание и оценка
print("\nОценка модели...")
y_pred = stacking_model.predict(X_test)

print("\nРезультаты стекинга:")
print(f"MAE: {mean_absolute_error(y_test, y_pred):.4f}")
print(f"R2: {r2_score(y_test, y_pred):.4f}")

# 5. Сравнение с базовыми моделями
print("\nСравнение с базовыми моделями:")

models = {
    'XGBoost': xgb_model.fit(X_train, y_train),
    'CatBoost': cat_model.fit(X_train, y_train),
    'LightGBM': lgbm_model.fit(X_train, y_train)
}

for name, model in models.items():
    y_pred = model.predict(X_test)
    print(f"\n{name}:")
    print(f"MAE: {mean_absolute_error(y_test, y_pred):.4f}")
    print(f"R2: {r2_score(y_test, y_pred):.4f}")
```

```python
# Получаем предсказания всех базовых моделей
model_predictions = []
model_names = []
for name, model in stacking_model.named_estimators_.items():
    preds = model.predict(X_test)
    model_predictions.append(preds)
    model_names.append(name)

model_predictions = np.array(model_predictions)

# Рассчитываем важность моделей через их веса в мета-модели
if hasattr(stacking_model.final_estimator_, 'coef_'):
    model_weights = np.abs(stacking_model.final_estimator_.coef_)
    model_weights /= model_weights.sum()

    # Сортируем веса и имена моделей по возрастанию
    sorted_indices = np.argsort(model_weights)
    sorted_model_weights = model_weights[sorted_indices]
    sorted_model_names = [model_names[i] for i in sorted_indices]

    plt.figure(figsize=(10, 5))
    bars = plt.barh(sorted_model_names, sorted_model_weights)
    plt.title('Вклад базовых моделей (по весам мета-модели)')
    plt.xlabel('Относительный вес')
    plt.tight_layout()

    # Добавляем значения на столбцы
    for bar in bars:
        width = bar.get_width()
        plt.text(width, bar.get_y() + bar.get_height()/2.,
                f'{width:.2f}',
                ha='left', va='center')

    plt.tight_layout()
    plt.show()
```

## Локальная интерпретация

```python
# 1. Подготовка данных для интерпретации
test_predictions = pd.DataFrame(
    stacking_model.transform(X_test),
    columns=[name for name, _ in stacking_model.estimators]
)

# 2. Создаем explainer для мета-модели
explainer = shap.Explainer(
    stacking_model.final_estimator_.predict, 
    test_predictions,
    feature_names=test_predictions.columns
)

# 3. Выбираем несколько случайных наблюдений для анализа
# Используем индексы от 0 до длины X_test, чтобы избежать проблем с индексацией
sample_indices = np.random.choice(len(X_test), size=3, replace=False)

# 4. Анализ каждого наблюдения
for i, idx in enumerate(sample_indices):
    print(f"\nНаблюдение {i+1} (Индекс в массиве: {idx})")
    
    # Получаем фактические значения, используя iloc для безопасного доступа
    if hasattr(y_test, 'iloc'):
        actual_value = y_test.iloc[idx]
    else:
        actual_value = y_test[idx]
    print(f"Фактическое значение: {actual_value:.2f}")
    
    # Получаем предсказание
    if isinstance(X_test, (pd.DataFrame, pd.Series)):
        prediction = stacking_model.predict(X_test.iloc[idx:idx+1])[0]
    else:
        prediction = stacking_model.predict(X_test[idx:idx+1])[0]
    print(f"Предсказанное значение: {prediction:.2f}")
    
    # SHAP значения для текущего наблюдения
    shap_values = explainer(test_predictions.iloc[[i]].values)
    
    # Waterfall plot
    shap.plots.waterfall(shap_values[0], max_display=10)
```
